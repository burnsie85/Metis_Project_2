```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

```python
#webscraping for 2020 player salaries

url2010 = 'https://hoopshype.com/salaries/players/2009-2010/'
        
page2010 = urlopen(url2010).read()
soup2010 = BeautifulSoup(page2010)
table2010 = soup2010.find_all('table')
df2010 = pd.read_html(str(table2010))

#somehow, the above code returned a list of one element.  
#That one element is a pandas dataframe.
#Thus, I inserted the below code to generate a pandas dataframe.

df2010 = df2010[0]

df2010 = df2010.rename(columns={'2009/10': 'Salary', '2009/10(*)':'Inflation Adjusted Salary'})
df2010["Year"] = 2010

df2010
```

```python
#setting a new column.  Inflation isn't factored in for the current year, so, I just
#made them equal

df2020['Inflation Adjusted Salary'] = df2020['Salary']
df2020
```

```python
#creating a dataframe that includes all years

years = [df2010, df2011, df2012, df2013, df2014, df2015, df2016, df2017, df2018, df2019, df2020]
df_sal = pd.concat(years)
df_sal
```

```python
#reformating a column name

df_sal = df_sal.rename(columns={'Inflation Adjusted Salary': 'inflation_adjusted_salary'})
df_sal
```

```python
#filling in NaN

df_sal.inflation_adjusted_salary.fillna(df_sal.Salary, inplace=True)
df_sal
```

```python
#getting rid of columns that aren't needed

df_sal = df_sal[['Player', 'Salary', 'inflation_adjusted_salary', 'Year']].copy()
```
